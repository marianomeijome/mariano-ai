```python
import skimage
import numpy as np
from sklearn.model_selection import train_test_split
import os
import matplotlib.pyplot as plt
import cv2
```

```python
plt.imread("")
```

```python
data_dir = 'pixar128/'
color_imgs = skimage.io.ImageCollection(data_dir + '/diffuse/*.jpeg')
normal_imgs = skimage.io.ImageCollection(data_dir + '/normal/*.jpeg')
```

```python
sz=512
color_imgs = skimage.img_as_float([cv2.resize(color_imgs[x],(sz,sz)) for x in range(len(color_imgs))])
normal_imgs = skimage.img_as_float([cv2.resize(normal_imgs[x], (sz,sz)) for x in range(len(normal_imgs))])
```

```python
color_imgs.shape
```

```python
seed = np.random.randint(999)
np.random.seed(seed); np.random.shuffle(color_imgs)
np.random.seed(seed); np.random.shuffle(normal_imgs)
```

```python
color_train, color_test, normal_train, normal_test = train_test_split(color_imgs, normal_imgs, test_size=0.2)
```

```python
plt.imshow(color_train[0])
plt.show()
plt.imshow(normal_train[0])
plt.show()
```

```python
from keras.models import Model
from keras.layers import Input, Conv2D, AveragePooling2D, UpSampling2D, BatchNormalization
from keras.optimizers import Adam
```

```python
bs = 8
epochs = 100
```

```python
input_img = Input(shape=(sz, sz, 3))

ng = Conv2D(15, (6, 6), activation='relu', padding='same', use_bias=False)(input_img)
ng = BatchNormalization()(ng)
ng = MaxPooling2D((4, 4), padding='same')(ng)
ng = Conv2D(30, (6, 6), activation='relu', padding='same', use_bias=False)(input_img)
ng = BatchNormalization()(ng)
ng = MaxPooling2D((4, 4), padding='same')(ng)
ng = Conv2D(60, (6, 6), activation='relu', padding='same', use_bias=False)(ng)
ng = BatchNormalization()(ng)
ng = UpSampling2D((4, 4))(ng)
ng = Conv2D(30, (6, 6), activation='relu', padding='same', use_bias=False)(ng)
ng = BatchNormalization()(ng)
ng = Conv2D(15, (6, 6), activation='relu', padding='same', use_bias=False)(ng)
ng = BatchNormalization()(ng)
ng = Conv2D(3, (6, 6), activation='sigmoid', padding='same')(ng)
normal_generator = Model(input_img, ng)
normal_generator.compile(Adam(amsgrad=True), loss='msle')
```

```python
normal_generator.summary()
```

```python
normal_generator.fit(color_train, normal_train,
                     validation_data = (color_test, normal_test),
                     batch_size=bs,
                     epochs=epochs)
```

```python
scores = normal_generator.evaluate(color_test, normal_test, verbose=0)
print("Baseline Error: %.4f%%" % (scores))
```

```python
def test_net_with_random_image():
    idx = int(np.random.rand(1)*len(color_test))
    y_legit = skimage.img_as_float(normal_test[idx]).reshape(1,sz,sz,3)
    scipy.misc.imsave('og.jpg', y_legit.reshape(sz,sz,3))
    y_fake = normal_generator.predict(skimage.img_as_float(color_test[idx]).reshape(1,sz,sz,3))
    scipy.misc.imsave('gen.jpg', y_fake.reshape(sz,sz,3))
    plt.imshow(y_legit.reshape(sz,sz,3))
    plt.show()
    plt.imshow(y_fake.reshape(sz,sz,3))
    plt.show()
```

```python
test_net_with_random_image()
```
